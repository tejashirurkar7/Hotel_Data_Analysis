```python
aggregated_bookings.shape
```

**Data Transformation**

```python
# Creating Occupancy Percentage Column

aggregated_bookings.head()
```

```python
aggregated_bookings['occ_pct'] = aggregated_bookings.apply(lambda row:row["successful_bookings"] / row["capacity"], axis = 1)
```

```python
aggregated_bookings.head()
```

```python
aggregated_bookings['occ_pct'] = aggregated_bookings['occ_pct'].apply(lambda x: round(x*100, 2))
aggregated_bookings.head()
```

**Insights Generation**

**1. What is an average occupancy rate in each of the room categories?**

```python
aggregated_bookings.head()
```

```python
aggregated_bookings.groupby("room_category")['occ_pct'].mean()
```

```python
# To get more clarity that what type of rooms it is 

merge_agg_rooms = pd.merge(aggregated_bookings, rooms, left_on="room_category" , right_on="room_id")
merge_agg_rooms.head()
```

```python
merge_agg_rooms.drop("room_id",axis = 1, inplace = True)
merge_agg_rooms.head()
```

```python
merge_agg_rooms.groupby("room_class")["occ_pct"].mean()
```

**2. What is Average Occupancy rate per city**

```python
hotels.head()
```

```python
merge_agg_rooms_hotels = pd.merge(merge_agg_rooms,hotels, on="property_id")
merge_agg_rooms_hotels.head()
```

```python
merge_agg_rooms_hotels.groupby("city")["occ_pct"].mean()
```

**3. When was the Occupancy better? Weekend or Weekday?**

```python
date.head()
```

```python
merge_agg_rooms_hotels_date = pd.merge(merge_agg_rooms_hotels,date, left_on="check_in_date", right_on="date")
merge_agg_rooms_hotels_date.head()
```
